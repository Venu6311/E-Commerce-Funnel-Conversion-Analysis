```python
## Create Time-Based Features

"""After converting the `event_time` column to the `datetime` data type,
additional time-based features were extracted to support exploratory
analysis and funnel analysis.

The extracted features include:

- **Hour:** To analyze customer activity throughout the day.
- **Weekday:** To identify purchasing patterns across different days of
  the week.

These features were later used in SQL analysis and Power BI dashboards
to identify peak customer activity and conversion trends."""
```

```python
df['event_time'] = pd.to_datetime(df['event_time'])
```

```python
df.info()
```

```python
## Remove Duplicate Records

"""Duplicate rows can introduce bias into the analysis by counting the same
customer interaction multiple times. Therefore, duplicate records were
identified and removed to ensure that each event represents a unique
customer interaction.

This step improves the accuracy and reliability of subsequent analyses,
including event distribution, conversion rates, and customer behavior
metrics."""


df.duplicated().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
## Remove Invalid Price Records

"""Rows containing negative product prices were removed because they do not
represent valid purchase transactions. Negative prices typically indicate
data entry errors, adjustments, refunds, or accounting corrections rather
than actual customer purchases.

Removing these records ensures that all analyses are performed using
valid transaction data and prevents misleading business insights."""


df=df[df['price']>=0]
```

```python
## Remove Records with Missing User Sessions

"""Rows containing missing values in the `user_session` column were removed
because each session represents a customer's shopping journey through
the platform.

Session identifiers are essential for accurately tracking customer
behavior, analyzing funnel progression, and calculating session-level
metrics. Records without a valid session ID cannot be reliably linked to
a customer journey and may lead to inaccurate analytical results.

Removing these records helps ensure the integrity and reliability of the
subsequent analysis."""
```

```python
df['user_session'].isnull().sum()
```

```python
df=df[df['user_session'].notnull()]
```

```python
df[df['user_session'].isnull()]
```

```python
## Remove High Missing Value Column

"""The `category_code` column contained approximately 98% missing values,
making it unsuitable for meaningful analysis.

Since the column provided very limited information and would not
contribute to business insights, it was removed from the dataset. This
helps reduce unnecessary data while improving the overall quality and
efficiency of the analysis."""
```

```python
df=df.drop(columns=['category_code'])
```

```python
## Data Cleaning Summary

"""The following preprocessing steps were performed to improve data quality:

- Removed duplicate records.
- Removed rows containing invalid (negative) product prices.
- Dropped the `category_code` column due to excessive missing values.
- Validated the cleaned dataset before exporting it for SQL analysis.

The cleaned dataset was then exported and used consistently throughout
the SQL analysis and Power BI dashboard development."""
```

```python
df.columns
```

```python
## Feature Engineering

"""To support time-based exploratory analysis and funnel analysis, additional
features were extracted from the `event_time` column.

The following columns were created:

- **date:** Represents the calendar date of each customer interaction,
  enabling daily trend analysis and time-series reporting.

- **hour:** Represents the hour of the day when the customer interaction
  occurred. This enables analysis of hourly customer activity and
  conversion trends.

- **day:** Represents the calendar day of the month, allowing analysis
  of daily customer activity patterns.

- **weekday:** Represents the day of the week (e.g., Monday, Tuesday),
  which helps identify customer behavior and purchasing trends across
  different weekdays.

- **month_name:** Represents the month in which the event occurred,
  supporting monthly trend analysis. Although this dataset contains data
  for October 2019, creating this feature improves the reusability of
  the preprocessing pipeline for future datasets containing multiple
  months.

These engineered features were later used in SQL analysis and Power BI
to analyze customer activity, conversion rates, and behavioral trends
over different time periods."""
```

```python
df['Date']=df['event_time'].dt.date
```

```python
df['Month_name']=df['event_time'].dt.month_name()
```

```python
df['Day']=df['event_time'].dt.day
```

```python
df['Weekday']=df['event_time'].dt.day_name()
```

```python
df['Hour']=df['event_time'].dt.hour
```

```python
## Export Cleaned Dataset
df.to_csv("cleaned_ecommerce.csv", index=False)
```
